```python
import torch
from PIL import Image
from transformers import AutoProcessor, AutoModelForVision2Seq
from transformers.image_utils import load_image
```

```python
DEVICE = "cpu" if torch.cuda.is_available() else "cpu"
print(f"Using {DEVICE}")

# Load images
image = Image.open("/home/dotronghiep/Documents/Research/VLM_Robot/0002_c1s1_069056_02.jpg")
```

```python
# Initialize processor and model
processor = AutoProcessor.from_pretrained("/home/dotronghiep/Documents/Research/VLM_Robot/smolvlm/local_processor_directory")
model = AutoModelForVision2Seq.from_pretrained(
    "/home/dotronghiep/Documents/Research/VLM_Robot/smolvlm/local_model_directory",
    torch_dtype=torch.bfloat16,
    _attn_implementation="flash_attention_2" if DEVICE == "cuda" else "eager",
).to(DEVICE)
```

```python
# Create input messages
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image"},
            {"type": "text", "text": "Can you describe images?"}
        ]
    },
]
```

```python
# Prepare inputs
prompt = processor.apply_chat_template(messages, add_generation_prompt=True)
inputs = processor(text=prompt, images=image, return_tensors="pt")
inputs = inputs.to(DEVICE)
```

```python
generated_ids = model.generate(**inputs, max_new_tokens=500)
```

```python
generated_texts = processor.batch_decode(
    generated_ids,
    skip_special_tokens=True,
)
```
